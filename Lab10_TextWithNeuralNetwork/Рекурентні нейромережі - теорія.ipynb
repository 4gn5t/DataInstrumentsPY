# Рекурентні нейромережі.

Рекурентні нейронні мережі (recurrent neural network, RNN) - це клас нейронних мереж, у яких з'єднання між вузлами мережі можуть створити цикл, дозволяючи передавати вихід з деяких вузлів на вхід інших вузлів, щоб створити ефект повторюваності. Вони можуть використовувати свій внутрішній стан (пам'ять) щоб обробляти послідовності входів різної довжини. <br>
Послідовності - це дані, які приходять у певному порядку.

## Шар рекурентної мережі.



Вузли, які є функціями входів з попередніх шарів також відомі як клітини пам'яті.



## Архітетура RNN.
Існують різні архітектури побудови RNN, ось приклад основних:
1. Sequence to Sequence (Many to Many).



У такій моделі на вхід подається послідовність, та на виході отримується послідовність.
2. Sequence to Vector (Many to One).



У такій моделі на вхід подається послідовність, а на виході отримується одне значення.
3. Vector to Sequence (One to Many).



У такій моделі на вхід подається одне значення, а на виході отримується послідовність.
<br>
Базові RNN мають недолік - вони можуть 'запам'ятати' тільки попередні входи. У них забуваються старі дані і враховуються тільки попередні t-1 виходи.

## LSTM - Long Short-Term Memory.
LSTM - це вузли, які були розроблені щоб вирішити проблеми, описані вище.
Модель нейрона LSTM:



* Forget Gate - вирішує, яку інформацію з попередніх одиниць пам'яті забувати;
* Input Gate - вирішує, яку інформацію з попередніх нейронів приймати;
* Update Gate - оновлює пам'ять;
* Output Gate - дає певний вихід з нейрона.
Gate, або ворота - це елемент, який вибірково пропускає інформацію. Математично ворота можна розглядати як функцію сигмоїда. Результати будть в межах між 0 та 1, при чому 1 будуть пропускатись.
* $x_t$ - вхід у момент часу t;
* $h_t$ - короткотривала пам'ять у момент часу t;
* $C_t$ - довготривала пам'ять у момент часу t.
<br><br>Працює цей нейрон так:
1. Створюється шар забування. Контекст цього шару - чим ближче отримане значення до 0, тим більша ймовірність, що його можна забути. Чим ближче отримане значення до 1, тим більша ймовірність, що його треба запам'ятати; 
2. Вирішується, яка інформація має зберегтись. Цей крок складається з двох етапів. Спочатку через сигмоїд вирішується, яка інформація буде оновлена. Потім, через гіперболізований тангенс, створюється вектор новіших значень кандидатів, які певним чином отримують стан нейрона;
3. Оновлюється стан пам'яті. Тут забуваються ті значення, які перший шар вирішив забути і потім додаються нові значення, які шар з кроку 2 вирішив залишити;
4. Вирішується, що подавати на вихід. Вихід нейрона залежить від оновленого стану нейрона, але відфільтрованого. Спочатку працює сигмоїд, який вирішує, яку інформацію буде подано на вихід. Потім стан нейрона пропускається через гіперболізований тангенс (всі значення попадають у проміжок між 1 та -1). Отримані значення перемножуються, щоб отримувати тільки ту інформацію, яка є необхідною.
