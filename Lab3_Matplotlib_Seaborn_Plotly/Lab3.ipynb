# MatPlotLib.

Наступне поле містить ініціалізацію даних, які необхідні для виконання завдань. Це поле необхідно виконати, але ніяких змін робити не треба.

```python
import numpy as np
x = np.arange(-10, 11)
y = x**2
q = x**3
```

#### Імпортуйте бібліотеку 'matplotlib' та додайте поле '%matplotlib inline'.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Побудуйте графік, для даних x та y. Задайте назву осей та графіку.

```python
# ваш код тут
plt.plot(x, y, label='x^2')
plt.title('Парабола')
```

#### Побудуйте дві системи координат, одна з яких буде мати параметри розмірів/розташування [0, 0, 1, 1] а інший [0.4, 0.5, 0.2, 0.2].

```python
# ваш код тут

fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.4, 0.5, 0.2, 0.2])
```

#### Додайте графіки до побудованих систем координат та покажіть їх на екрані. Перший графік для даних x i y, другий графік для даних x i q.

```python
# ваш код тут
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.4, 0.5, 0.2, 0.2])

axes1.plot(x, y, 'r')
axes2.plot(x, q, 'g')

```

#### Створіть два підполотна. У них побудуйте графік (x,y) та графік (x,q). Використайте різні параметри для кольору, товщини та стилю ліній графіків.

```python
# ваш код тут
figure, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r--')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[1].plot(x, q, 'g-.')
axes[1].set_xlabel('x')
axes[1].set_ylabel('q')
```

# Plotly.

Продовжуйте роботу з набором даних "Титанік".
Імпортуйте необхідні бібліотеки.

```python
import plotly.graph_objects as go
import plotly.express as px
```

```python
# ваш код тут
fig = px.histogram(titanic, x='class')
fig.show()
```

```python
# ваш код тут
titanic_numeric = titanic.select_dtypes(include=[np.number])
fig = px.imshow(titanic_numeric.corr(), title='Genre Popularity by Rating', color_continuous_scale='Viridis')
fig.show()

```

Як вхідні дані для мапи тепла використайте дані про кореляцію даних у наборі даних "Титанік".

```python
# ваш код тут
genre_counts = titanic.groupby('embark_town').size()
fig = go.Figure(data=[go.Pie(labels=genre_counts.index, values=genre_counts.values)])
fig.show()
```

```python
# ваш код тут
class_mapping = {'First': 1, 'Second': 2, 'Third': 3}
titanic['class_num'] = titanic['class'].map(class_mapping)

fig = go.Figure(data=[go.Scatter3d(
    x=titanic['age'],
    y=titanic['fare'],
    z=titanic['class_num'],
    mode='markers',
    marker=dict(
        size=12,
        color=titanic['survived'],                
        colorscale='Viridis',   
        opacity=0.8
    )
)])
fig.update_layout(title='Movie Ratings Over Time')
fig.show()
```
