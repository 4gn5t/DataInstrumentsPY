```python
# ваш код тут
dates = pd.date_range('2020-01-01', periods=100)
df = dd.from_pandas(pd.DataFrame({'date': dates, 'value': np.random.randint(0, 101, size=100)}), npartitions=4)
df.compute()
```

Виберіть одне з завдань з списку нижче та виконайте його використовуючи спочатку multiprocessing, а потім joblib.

Початковий масив даних створіть з допомогою NumPy, його розмір має бути не менше 100000 елементів. Якщо для виконання завдання потрібен додатковий параметр, його треба придумати самостійно. 

1. Сортування великого масиву: Розбити масив на частини та відсортувати кожну частину в окремому процесі. Потім злити відсортовані частини.
2. Сортування за парністю: Розділити масив на два списки: парні та непарні числа. Відсортувати кожен список окремо і з'єднати їх.
3. Паралельне обчислення статистик: Обчислити середнє, медіану, моду, дисперсію для різних частин великого масиву паралельно.
4. Паралельне застосування функцій: Застосувати задану математичну функцію (наприклад, синус, косинус) до кожного елемента масиву паралельно.
5. Паралельна фільтрація даних: Відфільтрувати елементи масиву за певним критерієм (наприклад, вибрати всі елементи, більші за задане число) паралельно.
6. Паралельна генерація випадкових чисел: Згенерувати велику кількість випадкових чисел різних розподілів (нормальний, рівномірний) паралельно.
7. Паралельне обчислення матричних операцій: Розбити велику матрицю на підматриці та виконати матричне множення або додавання підматриць в окремих процесах. Потім зібрати результат у єдину матрицю.
8. Паралельне підвищення ступеню матриці: Розбити матрицю на блоки та підвищити кожен блок до степеня паралельно. Потім об’єднати результат.
9. Пошук мінімуму та максимуму: Паралельно знайти мінімум та максимум для різних частин великого масиву.
10. Паралельне знаходження суми: Розділити масив на частини і паралельно обчислити суму елементів кожної частини. Потім підсумувати результати.
11. Паралельне піднесення до квадрату: Паралельно піднести кожен елемент великого масиву до квадрату, розділивши масив на частини.
12. Паралельна перевірка простоти чисел: Паралельно перевіряти, які числа у масиві є простими, і вивести їх у новий масив.
13. Паралельна зміна масштабу масиву: Паралельно масштабувати значення елементів масиву за допомогою множення на фіксований коефіцієнт.
14. Паралельне знаходження індексів: Розділити масив на частини і паралельно знайти індекси елементів, які відповідають певному критерію (наприклад, більше заданого числа).
15. Паралельне знаходження кількості позитивних чисел: Розбити масив на частини і паралельно підрахувати кількість позитивних елементів у кожній частині. Після цього об'єднати результати.
